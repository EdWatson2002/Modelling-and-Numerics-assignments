```python
import numpy as np
from numpy.linalg import norm
from numpy import sin,cos,pi,exp,array, exp, zeros, sqrt
from IPython.display import display
from matplotlib import pyplot as plt
import pandas as pd

def npPrint(a, f={'float_kind':lambda x: "%.5e" % x}):
    print( np.array2string(a, formatter=f) )

def Q11( f,Df, t0,y0, h):
    f0 = f(t0,y0)
    f1 = f(t0+h, y0+h*f0)
    return y0 + h/2*(f0+f1)

def forwardEuler(f,Df,  tn,yn, h):
    return yn + h * f(tn,yn)

def computeEocs( herr ):
    # herr = ( (h1,e1),(h2,e2),...,(hm,em) )
    m = len(herr)
    eocs = np.zeros( m-1 )
    for i in range(m-1):
        eocs[i] = np.log(herr[i+1,1]/herr[i,1] ) /\
                  np.log(herr[i+1,0]/herr[i,0] );
    return eocs


def evolve(phi,f,Df, t0,y0, T,N):
    h = T/N # could also be h=(T-t0)/N
    y = np.zeros([N+1, len(y0)])
    t = np.zeros( N+1 )
    y[0] = y0
    t[0] = 0   # could be t[0]=t0
    for i in range(N):
        y[i+1] = phi(f,Df, t[i],y[i], h)
        t[i+1] = t[i]+h
    return t,y

from numpy.linalg import solve
def newton(F,DF,x0,eps,K):
    k  = 0
    x  = x0.copy().astype(np.float)  # note: with x=x0 changes to x also changes to x0 with numpy arrays
    Fx = F(x)
    while Fx.dot(Fx) > eps*eps and k<K:
        x[:] -= solve(DF(x), Fx)  # don't construct a new vector in each step - they could be large
        Fx = F(x)
        k += 1
    return x,k

def backwardEuler( f,Df, t0,y0, h):
    # not good style but reduces polution of the global namespace when running the tests
    from numpy import eye
    F =  lambda delta: delta-f(t0+h,y0+h*delta)
    DF = lambda delta: eye(len(y0))-h*Df(t0+h,y0+h*delta)
    delta,_ = newton(F,DF,f(t0+h,y0),h*h*1e-5,1000)
    return y0+h*delta

def crank(f, Df, t0, y0, h):
    F =  lambda delta: delta-f(t0+h,y0+(h/2)*(f(t0,y0) + delta))
    DF = lambda delta: np.eye(len(y0))-(h/2)*Df(t0+h,y0+(h/2)*(f(t0,y0) + delta))
    delta,_ = newton(F,DF,f(t0+h,y0),h*h*1e-5,1000)
    return y0 + (h/2)*(f(t0,y0) + delta)

def experiment(stepper, f,Df,T,Y, M,N0):
    herr = zeros( [M,2] )
    for m in range(M):
        N = N0*2**m
        try:
            t,y = evolve( stepper, f,Df, 0,Y(0),T, N)
            herr[m] = [T/N, max( norm(Y(t).transpose()-y,axis=1) )]
        except ValueError: # method not stable for given 'h'
            print("Error:", stepper,N,T/N,"failed: some NaN value produced!")
            herr[m] = [T/N, np.nan]
        except AssertionError: # assert in dirk was triggered (failed Newton)
            print("Error:", stepper,N,T/N,"failed: Newton function failed!")
            herr[m] = [T/N, np.nan]
            
    eoc = computeEocs( herr )
    eoc = np.hstack( [ [np.nan],eoc ] )
    res = np.hstack( [herr, np.atleast_2d(eoc).T] )
    return res

def compareErrors(results):
    # columns of table
    simul = list(results.keys())
    columns=['h'] + [s for n in simul for s in [f'{n}-error', f'{n}-eoc']]
    # values in tablea
    keys = {columns[0]: results[simul[0]][:,0]} # all results are assumed to␣
    styles = {columns[0]: '{:.4e}'}
    for i, k in enumerate(simul):
        keys[columns[2*i+1]] = results[k][:,1] # errors
        styles[columns[2*i+1]] = '{:.6e}'
        keys[columns[2*i+2]] = results[k][:,2] # eocs
        styles[columns[2*i+2]] = '{:.3f}'

    table = pd.DataFrame(keys, index=range(results[simul[0]].shape[0]),columns=columns)
    
    display( table.style.format(styles) )
    for i,k in enumerate(simul):
        plt.loglog(results[k][:,0],results[k][:,1],marker='o',label=k)
    plt.legend()
    plt.grid(True)
    plt.xlabel("step size h")
    plt.ylabel("Maximum error over time")
    plt.savefig("Q2_compareErr.pdf", format="pdf", bbox_inches="tight")
    plt.show()
```

```python
def dirk(f,Df, t0,y0, h, alpha ,beta, gamma):
    s = len(alpha)      # m-stage rk method
    dim_y = len(y0)
    k = zeros((s,dim_y))
    for i in range(s):

        sum_less_k = zeros(dim_y)
        for j in range(i):
            sum_less_k =  sum_less_k + beta[i][j]*k[j]
        
        newtonF = lambda d: d - f(t0 + alpha[i]*h , y0 + h*(sum_less_k + beta[i][i]*d))
        newtonDF = lambda d: np.eye(dim_y) - h*beta[i][i]*Df(t0 + alpha[i]*h , y0 + h*(sum_less_k + beta[i][i]*d))
        
        k[i],_ = newton(newtonF,newtonDF,f(t0,y0),1e-15,100)
    
    phi = zeros(dim_y)
    for i in range(s):
        phi = phi + gamma[i]*k[i]
    return y0 + h*phi

def taylor(f,Df,dtdf, t0,y0, h):
    return y0 + h*f(t0,y0) + ((h**2)/2)*(dtdf(t0,y0) + Df(t0,y0).dot(f(t0,y0)))
```

# Discussion

First, we check order conditions. Since $\gamma = (\frac{1}{2},\frac{1}{2})$ for all 3 methods, $\sum_{i = 1}^{m} \gamma_i = 1$, so all methods have $p\geq1$.

For $d = 0$, the butcher tableau is 
$$\begin{array}
{c|cccc}
0\\
1 & 1\\
\hline
& \frac{1}{2} &\frac{1}{2} 
\end{array}$$

First we check the order condition for $p\geq3$. $\sum_{i = 1}^{m} \alpha_i^{2} \gamma_i = \frac{1}{2} \neq \frac{1}{3}$, so the condition is not met, thus we check for $p\geq2$.$\sum_{i = 1}^{m} \alpha_i \gamma_i = \frac{1}{2}$, so the condition is met, thus for $d=0$, the method has order of consistency $p=2$

Next, the Butcher tableau for $d = \frac{1}{4}$ is
$$\begin{array}
{c|cccc}
\frac{1}{4} & \frac{1}{4}\\
\frac{3}{4} & \frac{1}{2} & \frac{1}{4}\\
\hline
& \frac{1}{2} &\frac{1}{2} 
\end{array}$$

We first check the condition for $p\geq3$. $\sum_{i = 1}^{m} \alpha_i^{2} \gamma_i = \frac{5}{16}$ so the condition is not met. We check $p\geq2$. $\sum_{i = 1}^{m} \alpha_i \gamma_i = \frac{1}{2}$, thus for $d=\frac{1}{4}$, the method has order of consistency $p=2$

Finally, the Butcher tableau for $d = \frac{1}{2} + \frac{\sqrt{3}}{6}$ is 
$$\begin{array}
{c|cccc}
\ \frac{1}{2}+\frac{\sqrt{3}}{6} & \frac{1}{2}+\frac{\sqrt{3}}{6}\\
\ \frac{1}{2}-\frac{\sqrt{3}}{6} & \frac{1}{2}+\frac{\sqrt{3}}{6} & \frac{1}{2}+\frac{\sqrt{3}}{6}\\
\hline
& \frac{1}{2} &\frac{1}{2} 
\end{array}$$

We check the conditions for $p\geq3$ first. $\sum_{i = 1}^{m} \alpha_i^{2} \gamma_i = \frac{1}{3}$  and 

$$\sum_{i=1}^{m}\sum_{j=1}^{m}\gamma_i\beta_{i,j}\alpha_j = d-d^2 = \frac{1}{6}$$
So this method has order of consistency $p=3$.

```python
from scipy . special import ellipj , ellipk

class hamiltonian:

    x0 = 0
    p0 = 0
    omega0 = 0

    def __init__(self, x0, p0, omega0):
        self.x0 = x0
        self.p0 = p0
        self.y0 = array([x0,p0])
        self.omega0 = omega0


    def Y(self,t):
        k = np.sin(self.x0 /2) # x0: initial condition
        K = ellipk (k*k)
        sn , cn , dn , _ = ellipj (K- self.omega0 *t, k*k)
        x = 2* np.arcsin (k*sn)
        p = 2/ sqrt (1 -(k*sn )**2) * cn*dn * k*( - self.omega0 )
        return array ([x,p])

    def f(self,t,y):
        return array([y[1], -(self.omega0**2)*sin(y[0])])

    def Df(self,t,y):
        return array([[0,1], [-(self.omega0**2)*cos(y[0]), 0]])
```

```python
first_y0 = hamiltonian(0.05*pi,0,4)

params = (first_y0.f,first_y0.Df,100,first_y0.Y,8,500)

exp1 = experiment(fixed_dirk(0.25), *params)
exp2 = experiment(fixed_dirk(0.5 + sqrt(3)/6), *params)
                                        
compareErrors({"d = 1/4": exp1,  "d = 1/2 + sqrt(3)/6": exp2})
```

```python
from scipy.linalg import expm

def eDirk( F,DF, t0,y0, h, alpha,beta,gamma, exphMs, exphMsInv, exphM):
    s = len(alpha)      # m-stage rk method
    dim_y = len(y0)
    k = zeros((s,dim_y))
    for i in range(s):

        sum_less_k = zeros(dim_y)
        for j in range(i):
            sum_less_k =  sum_less_k + beta[i][j]*k[j]
        
        newtonF = lambda d: d - exphMsInv[i].dot(F(t0 + alpha[i]*h , exphMs[i].dot(y0 + h*(sum_less_k + beta[i][i]*d))))
        newtonDF = lambda d: np.eye(dim_y) - h*beta[i][i]*exphMsInv[i].dot( DF(t0 + alpha[i]*h ,exphMs[i].dot( y0 + h*(sum_less_k + beta[i][i]*d))))
        
        k[i],_ = newton(newtonF,newtonDF,F(t0,y0),1e-15,100)
    
    phi = zeros(dim_y)
    for i in range(s):
        phi = phi + gamma[i]*k[i]
    return exphM.dot(y0 + h*phi)

def experiment2(stepper, f,Df,T,Y, M,N0, mat, alpha, beta, gamma):
    herr = zeros( [M,2] )
    s = len(alpha)
    size = len(Y(0))
    for m in range(M):
        N = N0*2**m
        h = T/N

        exphM = expm(h*mat)

        exphMs = zeros((s,size,size))
        for i in range(s):
            exphMs[i] = expm(h*alpha[i]*mat)

        exphMsInv = zeros((s,size,size))
        for i in range(s):
            exphMsInv[i] = np.linalg.inv(expm(h*alpha[i]*mat))

        fixed_stepper = functools.partial(stepper,
                                alpha = alpha, 
                                beta = beta, 
                                gamma = gamma,
                                exphM = exphM,
                                exphMs = exphMs,
                                exphMsInv = exphMsInv
                                )

        try:
            t,y = evolve( fixed_stepper, f,Df, 0,Y(0),T, N)
            herr[m] = [T/N, max( norm(Y(t).transpose()-y,axis=1) )]
        except ValueError: # method not stable for given 'h'
            print("Error:", stepper,N,T/N,"failed: some NaN value produced!")
            herr[m] = [T/N, np.nan]
        except AssertionError: # assert in dirk was triggered (failed Newton)
            print("Error:", stepper,N,T/N,"failed: Newton function failed!")
            herr[m] = [T/N, np.nan]
            
    eoc = computeEocs( herr )
    eoc = np.hstack( [ [np.nan],eoc ] )
    res = np.hstack( [herr, np.atleast_2d(eoc).T] )
    return res
```

```python
class expHamiltonian:

    x0 = 0
    p0 = 0
    w0 = 0

    def __init__(self, x0, p0, omega0):
        self.x0 = x0
        self.p0 = p0
        self.y0 = array([x0,p0])
        self.w0 = w0


    def Y(self,t):
        k = np.sin(self.x0 /2) # x0: initial condition
        K = ellipk (k*k)
        sn , cn , dn , _ = ellipj (K- self.w0 *t, k*k)
        x = 2* np.arcsin (k*sn)
        p = 2/ sqrt (1 -(k*sn )**2) * cn*dn * k*( - self.w0 )
        return array ([x,p])

    def F(self,t,y):
        return array([0, w0*w0*y[0] - w0*w0*sin(y[0])])

    def DF(self,t,y):
        return array([[0,0],[w0*w0 - w0*w0*cos(y[0]), 0]])

w0 = 4
mat = array([[0,1],[-w0*w0, 0]])

first_y0 = expHamiltonian(0.05*pi,0,w0)

params = lambda d: (eDirk, first_y0.F,first_y0.DF,100,first_y0.Y,8,500, mat,
                    array([d,1-d]), 
                    array([[d,0],[1-2*d,d]]), 
                    array([0.5,0.5]) )

exp1 = experiment2(*params(0.25))
exp2 = experiment2(*params(0.5 + sqrt(3)/6))
                                        
compareErrors({"d = 1/4": exp1,  "d = 1/2 + sqrt(3)/6": exp2})
```

```python
second_y0 = expHamiltonian(0.6*pi,0,w0)

params = lambda d: (eDirk, second_y0.F,second_y0.DF,100,second_y0.Y,8,500, mat,
                    array([d,1-d]), 
                    array([[d,0],[1-2*d,d]]), 
                    array([0.5,0.5]) )

exp1 = experiment2(*params(0.25))
exp2 = experiment2(*params(0.5 + sqrt(3)/6))
                                        
compareErrors({"d = 1/4": exp1,  "d = 1/2 + sqrt(3)/6": exp2})
```
